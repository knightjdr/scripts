Generate a histrogram of BFDR for a SAINT file

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
saint_file = 'data/mito.txt'
```

```python
def read_saint(saint_file):
  columns = ['Bait', 'Prey', 'PreyGene', 'BFDR']
  df = pd.read_csv(saint_file, sep='\t', usecols=columns)
  return df

df = read_saint(saint_file)
```

```python
bins = np.arange(0, 1, 0.05)

df.plot.hist(by='BFDR', bins=bins)
plt.title('BFDR distribution')
plt.xlabel('BFDR')
plt.ylabel('Interactions')

stem = Path(saint_file).stem
outfile = f'output/{stem}.pdf'
plt.savefig(outfile)
```
